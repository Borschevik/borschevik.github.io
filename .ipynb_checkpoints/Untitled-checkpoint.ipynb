```python
class DTPTRansformer(TransformerMixin, BaseEstimator):
    
    def fit(self, json_path):
        return self
    
    def get_df_columns(self, column, condition):
        coord = df.target[self.df[column] != condition]
        return coord
        
    def get_months_coords(self, months):
        return self.df.loc[:, ['geometry', 'ndu']][self.df.index.month.isin(months)]
    
    def save_data(self, df, filename):
        return df.to_file(filename, driver='GeoJSON')
    
    def transform(self, json_path):
        parser = lambda date: pd.datetime.strptime(date, '%H:%M %Y.%m.%d')
        self.df = gpd.read_file(json_path)
        self.df.index = pd.to_datetime(self.df.Time + ' ' +self.df.date.astype(str))
        D = {'pog':[], 'znach_dor':[], 'ndu':[], 'change_org_motion':[], 'osv':[], 'sost_pokritiya':[]}
        
        for inf in self.df.infoDtp:
            D['pog'].append(' '.join(inf['s_pog']))
            D['znach_dor'].append(inf['dor_z'])
            D['ndu'].append(' '.join(inf['ndu']))
            D['change_org_motion'].append(inf['change_org_motion'])
            D['osv'].append(inf['osv'])
            D['sost_pokritiya'].append(inf['s_pch'])
        parse_naezd = lambda x: 'Наезд' if 'Наезд' in x else x
        self.df.DTP_V = self.df.DTP_V.apply(parse_naezd)
        self.df = pd.concat([self.df, pd.DataFrame(D, index=self.df.index)], axis=1)
        self.df=self.df[self.df.DTP_V.isin(['Наезд',
        'Столкновение',
        'Падение пассажира',
        'Опрокидывание',
        'Иной вид ДТП'])]
        self.df['Календарное время года'] = np.zeros(self.df.shape[0])
        self.df['Календарное время года'][self.df.index.month.isin([1,2])]  = 1
        self.df['Календарное время года'][self.df.index.month.isin([12])]= 1
        self.df['Календарное время года'][self.df.index.month.isin([3,4,5])] = 2
        self.df['Календарное время года'][self.df.index.month.isin([6,7,8])] = 3
        self.df['Календарное время года'][self.df.index.month.isin([9,10,11])] = 4
        self.df['Часть дня'] = np.zeros(self.df.shape[0])
        df1 = self.df.loc[self.df.between_time('08:00', '11:59').index.values] 
        df1['Часть дня'] = 'first'
        df2 = self.df.loc[self.df.between_time('12:00', '17:59').index.values] 
        df2['Часть дня'] = 'second'
        df3 = self.df.loc[self.df.between_time('18:00', '20:59').index.values] 
        df3['Часть дня'] = 'third'
        df4 = self.df.loc[self.df.between_time('21:00', '23:59').index.values] 
        df4['Часть дня'] = 'last'
        df5 = self.df.loc[self.df.between_time('00:00', '07:59').index.values] 
        df5['Часть дня'] = 'last'
        self.df = pd.concat([df1,df2,df3,df4,df5], axis=0)
        res = pd.concat([
        self.df.loc[:, ['Календарное время года', 'Часть дня'] ],
        self.df.loc[:, D.keys()]], axis=1
        )
        Y, X = pd.get_dummies(self.df.DTP_V, columns=self.df.DTP_V.unique()), res
        return Y, X

```

```python
def check_coors(df, names=['K_UCH','RAN','POG','K_TS'], name='dtp_corr', cat='DTP_V'):
    df = df.copy()
    df = pd.concat([df.loc[:, names],
    pd.get_dummies(df[cat])], axis=1)

    corr = df.corr()

    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

    f, ax = plt.subplots(figsize=(30, 30))
    sns.set(font_scale=5)
    cmap = sns.diverging_palette(320, 20, as_cmap=True)
    sns.set(font_scale=3)
    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

    plt.savefig("./images/{}.png".format(name), dpi=300)
```

```python
def optimal_eps(df):
    val = []
    for ep in range(500, 6000, 500):
        db = DBSCAN(eps=ep, algorithm='kd_tree', metric='euclidean')
        db.fit(np.matrix(np.column_stack((df.geometry.x, df.geometry.y))))
        val.append((ep, len(db.labels_[db.labels_ == -1])/len(db.labels_)))

    labels, ys = zip(*val)
    width = 1
    sns.set(font_scale=3)
    fig = plt.figure(figsize=(30, 30))                                                               
    ax = fig.gca() 
    ax.plot(labels, ys, width, color='red')
    plt.show()
```

```python
def draw_importance(columns, importances, top_k=20, save_path=None):
    order = np.argsort(importances)[-top_k:]
    objects = columns[order]
    labels = []
    for objec in objects:
        if '_' in objec:
            objec = objec.split('_')[-1]
        labels.append(objec[0:30] if len(objec) > 20 else objec)
    y_pos = np.arange(len(objects))
    performance = np.array(importances)[order]

    plt.figure(figsize=(30, 30))
    sns.set(font_scale=4)
    plt.barh(y_pos, performance, align='center', alpha=0.5)
    plt.yticks(y_pos, labels, )
    plt.xlabel('Важность')
    plt.title('Фактор')

    if save_path:
        plt.savefig(save_path)
    plt.show()
    return objects
```

```python
objects = draw_importance(X_train.columns, model.feature_importances_, save_path='./images/importance.png')
```
